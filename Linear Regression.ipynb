```python
import numpy as np
import pandas as pd
import torch
```

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()

boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names, dtype = "float32")
boston.head()
```

```python
boston['MEDV'] = boston_dataset.target
```

```python
boston.head()
```

```python
boston['MEDV'].dtypes
```

```python
input = boston.drop(["MEDV"], axis=1)
input.head()

input = torch.tensor(input.values)
```

```python
target = boston["MEDV"]
target.head()

target = torch.tensor(target.values)
```

```python
w = torch.randn(input.shape[0],input.shape[1], requires_grad=True)
b = torch.randn(input.shape[0], requires_grad=True)
```

```python
def model(x):
    return x @ w.t() + b
```

```python
prediction = model(input)
print(prediction)
```

```python
print(target)
```

```python
#Loss function

def mse(j1,j2):
    diff = j1-j2
    return torch.sum(diff**2)/diff.numel()
```

```python
loss = mse(prediction, target)
print(loss)
```

```python
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
```

```python
print(w)
print(b)
```
